```python
worksheet=gc.open('student_data').sheet1
rows = worksheet.get_all_values()

```

```python
dataset1 = pd.DataFrame(rows[1:], columns=rows[0])
dataset1 = dataset1.astype({'Input': 'int'})
dataset1 = dataset1.astype({'output': 'int'})
```

```python
dataset1.head()
```

```python
X = dataset1[['Input']].values
y_train = dataset1[['output']].values
```

```python
#x
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y_train, test_size=0.33, random_state=33)
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train1= scaler.transform(X_train)
```

```python
ai_brain = Sequential([
Dense(8,activation='relu'),
Dense(4,activation='relu'),
Dense(1)
])
```

```python
ai_brain.compile(optimizer= 'rmsprop', loss="mse")
```

```python
ai_brain.fit(X_train1,y_train,epochs=2000)
```

```python
loss_df = pd.DataFrame(ai_brain.history.history)
```

```python
loss_df.plot()
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_test1 = scaler.transform(X_test)
```

```python
ai_brain.evaluate(X_test1,y_test)
```
